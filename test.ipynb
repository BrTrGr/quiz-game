```python
size = 4

for i in range(size, 0, -1):
    if i == size:
        print("*" * ((size * 2) - 1))
    elif i == 1:
        print(" " * (size - i), "*")
    else:
        outer_space = " " * (size - i)
        inner_space = " " * ((i - 1) * 2 + 1)
        print(outer_space, "*", inner_space, "*")

for i in range(2, size + 1):
    if i == size:
        print("*" * ((size * 2) - 1))
    else:
        outer_space = " " * (size - i)
        inner_space = " " * ((i - 1) * 2 + 1)
        print(outer_space, "*", inner_space, "*")
```
