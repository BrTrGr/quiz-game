INTERACT WITH DOCUMENTS

w - create and write
with open("words.txt", "w") as f:
    for i in range(100):
       f.write("Python awesome!\n") 

a - append
with open("words.txt", "a") as f:
    for i in range(100):
       f.write("Python awesome!\n") 

r - read
with open("words.txt", "r") as f:
    t = f.read()
    print(t.split())
